### Defination:

    Matplotlib is one of the most popular python packages used for data visualization,.it is a cross-platform library for making 2D plots from data in arrays

```python
x = [5,2,7]
y = [2,16, 4]
plt.plot(x,y)
plt.title('Diagram')
plt.show()
```

### pyplot

```python
xpoints = np.array([0,6])
ypoints = np.array([0,9])
plt.plot(xpoints, ypoints)
plt.show()
```

### plotting without line

```python
xpoints = np.array([1,8])
ypoints = np.array([3,10])
plt.plot(xpoints, ypoints,'o')
plt.show()
```

### Multiple points

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,7,10])
plt.plot(xpoints, ypoints)
plt.show()
```

### Default points

```python
ypoints = np.array([3,8,1,10,5,7])
plt.plot(ypoints)
plt.show()
```

### Markers

```python
ypoints = np.array([3,8,1,10])
#plt.plot(ypoints,marker = 'o')
plt.plot(ypoints, marker = '*')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o',ms = 10,mec = 'r',mfc = 'g')
```

### Line plot

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,linestyle = 'dotted')
plt.show()
```

```python
plt.plot(ypoints,color = 'r')
plt.show()
```

```python
plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

### Subplot

```python
#plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(x,y)

#plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

### Barplot

```python
langs = ['Python','Javascript','Java','Php','Boostrap']
students = [5,40,20,15,12]
plt.bar(langs,students)
plt.show()
```

### Histogram plot

```python
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a,bins = [0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('number of students')
plt.show()
```
