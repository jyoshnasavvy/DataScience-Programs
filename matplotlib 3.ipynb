## setting limits tick labels

```python
import matplotlib.pyplot as plt
from matplotlib.figure import Figure
```

```python
x = [3, 1, 3]
y = [3, 2, 1]
fig = plt.figure (figsize = (5, 4))   #Creating a new figure with width=5 inchesand height=4inches
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])   #Creating first axes for the figure
ax.plot(x, y)                            #Adding the data to be p[lotted
ax.set_xlim(1, 2)
ax.set_xticklabels(("one", "two", "three", "four", "five", "six"))
plt.show()
```

##  Multiple plots

```python
fig = plt.figure(figsize =(5, 4))
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])   #Creating first axes for the figure
ax2 = fig.add_axes([0.5,0.5,0.3,0.3])          #Creating first axes for the figure
ax1.plot([5, 4, 3, 2, 1], [2, 3, 4, 5, 6])     #Adding the data to be plotted
ax2.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6])
plt.show()
```

```python
x = [3, 1, 3]
y = [3, 2, 1]
z = [1, 3, 1]
#Creating figure subplot
plt.subplot(121)
plt.plot(x, y)
#Adding second subplot
plt.subplot(121)
plt.plot(z, y)
```

```python
# Creating the figure and sunplots
#according the argument passed
fig, axes = plt.subplots(1, 2)
#plotting the data in the 1st subplot
axes[0].plot([1, 2, 3, 4], [1, 2, 3, 4])
#plotting the data in the 1st subplot only
axes[0].plot([1, 2, 3, 4], [1, 2, 3, 4])
#plotting the data in the 2nd subplot only
axes[1].plot([1, 2, 3, 4,],[1, 1, 1, 1])
```

```python
# data to display on plots
x = [3, 1, 3]
y = [3, 2, 1]
z = [1, 3, 1]
#Adding the subplots
axes1 = plt.subplot2grid((7, 1), (0, 0), rowspan = 2, colspan = 1)
axes2 = plt.subplot2grid((7, 1), (2, 0), rowspan = 2, colspan = 1)
axes3 = plt.subplot2grid((7, 1), (4, 0), rowspan = 2, colspan = 1)
# pltting the data
axes1.plot(x, y)
axes2.plot(x, y)
axes3.plot(x, y)
```
